```python
data["date"] = pd.to_datetime(data["date"],format = '%Y-%m-%d')

# here i used the to_datetime to convert the string datetimr into python datetime format
```

```python
data['year'] = data['date'].dt.year
data["month"] = data["date"].dt.month
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15,10))
plt.title("Temperature Change in Delhi Over the Years")
sns.lineplot(data = data, x='month', y='meantemp', hue='year')
plt.show()
```

```python
forecast_data = data.rename(columns = {"date": "ds",
                                       "meantemp": "y"})
print(forecast_data.head())


#docs about prophet


#first i am going to expalin what is prophet..........
# ok prophet is a python and r machine learning library use to
# forecasting or to deal with the time series model.
```

```python
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
model = Prophet()
model.fit(forecast_data)
forecasts = model.make_future_dataframe(periods = 100)
predictions = model.predict(forecasts)
plot_plotly(model, predictions)
```
